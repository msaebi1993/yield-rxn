```python
def reaction_classification(reaction_list,item_list):
    reaction_index=[]
    rm_list=[]
    for n,rec_eg in enumerate(reaction_list):
        for item in item_list:
            if rec_eg[1] is not None:
                
                factor = rec_eg[1].lower().find(item)
                if factor !=-1:
                    reaction_index.append([rec_eg[0],rec_eg[1]])
                    rm_list.append(n)
                    break
    new_list = [j for i, j in enumerate(reaction_list) if i not in rm_list]
    return new_list,reaction_index
```

```python
def get_smiles(index):
    
    if index == None:
        return None
    else:
        smiles = None
        name = None
        
        for mol in Molecule:
            if index==mol['@ID']:
                MOL = mol['MOLSTRUCTURE']
                NAME = mol['NAME']
                print(index)
                try:
                    for compounds in pcp.get_compounds(MOL, 'sdf') :
                        smiles= compounds.isomeric_smiles
                    if smiles ==None:
                        for compounds in pcp.get_compounds(NAME, 'name') :
                            smiles= compounds.isomeric_smiles
                        
                except:
                    continue
        return smiles
        
```

```python
def get_compound(n,entry,name):
    if entry is None:
        return None 
    else:
        compounds = []
        if type(entry) == list:
            for cp in entry:
                compounds.append(get_smiles(cp['@ID']))
        else:
            compounds.append(get_smiles(entry['@ID']))
        print(compounds)
        return compounds 
```
